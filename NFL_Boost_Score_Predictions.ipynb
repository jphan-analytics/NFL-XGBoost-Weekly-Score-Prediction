```python
#Filling in missing values
columns = ['playoff', 'importance', 'total_rating']
df.loc[:, columns] = df[columns].fillna(0)
```

```python
# Remove whitespaces to make tree later
columns_to_replace = ['qb1', 'qb2']

trimmed_df.loc[:, columns_to_replace] = trimmed_df[columns_to_replace].apply(lambda x: x.str.replace(' ', '_'))
```

```python
columns_to_drop = ['score1', 'score2','date']
X = trimmed_df.drop(columns=columns_to_drop)
```

```python
y1=trimmed_df['score1'].copy()
y2=trimmed_df['score2'].copy()
```

```python
X_encoded = pd.get_dummies(X, columns=['playoff',
                                       'team1',
                                       'team2',
                                       'qb1',
                                       'qb2'],
                          drop_first=True)        
X_encoded.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y1_train, y1_test, y2_train, y2_test = train_test_split(X_encoded, y1, y2, random_state=42, test_size=0.2)
```

```python
!pip install xgboost

import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
# Create and train the model for y1 scores with early stopping
model_y1 = xgb.XGBRegressor()
model_y1.fit(X_train,
            y1_train,
            verbose=True,
            early_stopping_rounds=10,
            eval_set=[(X_test, y1_test)]
        )

# Predict y1 scores
y1_prediction = model_y1.predict(X_test)

# Evaluate the model for y1 scores (using RMSE in this case)
rmse_y1 = mean_squared_error(y1_test, y1_prediction, squared=False)
print("RMSE for y1 scores:", rmse_y1)

# Create and train the model for y2 scores with early stopping
model_y2 = xgb.XGBRegressor()
model_y2.fit(X_train,
            y2_train,
            verbose=True,
            early_stopping_rounds=10,
            eval_set=[(X_test, y2_test)]
        )

# Predict y2 scores
y2_prediction = model_y2.predict(X_test)

# Evaluate the model for y2 scores (using RMSE in this case)
rmse_y2 = mean_squared_error(y2_test, y2_prediction, squared=False)
print("RMSE for y2 scores:", rmse_y2)
```

```python
#Tuning hyperparameters
from sklearn.model_selection import GridSearchCV

#Round1
# param_grid = {
#     'max_depth': [3, 4, 5],
#     'learning_rate': [0.1, 0.01, 0.05],
#     'gamma': [0, 0.25, 1.0],
#     'reg_lambda': [0.1, 1, 10],
#     'scale_pos_weight': [1, 3, 5]
# }
#Output: {'gamma': 0.25, 'learning_rate': 0.1, 'max_depth': 4, 'reg_lambda': 10, 'scale_pos_weight': 1}

#Round 2
param_grid = {
    'max_depth': [4,],
    'learning_rate': [0.1, 0.5, 1],
    'gamma': [0, 0.25],
    'reg_lambda': [10, 20, 100],
    'scale_pos_weight': [1]
}

grid_search = GridSearchCV(estimator=model_y1, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)

# Perform grid search with cross-validation
grid_search.fit(X_train, y1_train)

# Get the best parameters and best estimator from the grid search
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Predict y1 scores
y1_prediction = best_model.predict(X_test)

# Evaluate the best model for y1 scores (using RMSE in this case)
rmse_y1 = mean_squared_error(y1_test, y1_prediction, squared=False)
print("Best RMSE for y1 scores:", rmse_y1)
print("Best hyperparameters:", best_params)
```
